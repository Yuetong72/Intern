```python
import pandas as pd
import numpy as np
import os
import shutil
import warnings
warnings.filterwarnings('ignore')

shutil.rmtree("./output", ignore_errors=True)
os.makedirs("./output")

sheet = pd.read_excel('Region_Country.xlsx')
for i in range(0,len(sheet)):
    
    region = sheet.iloc[i,0]
    country = sheet.iloc[i,1] 
    print(i, country)
    raw_data = pd.read_excel(region, sheet_name = country)

#    df = raw_data.iloc[2:117,:]
#    df1_1 = df.iloc[1:,:]
#    df1_1.columns = df.iloc[0]

    df1_1 = raw_data.iloc[3:117,:]
#    df1_1 = df1_1.drop([0,1,2], axis =0)
    df1_1.columns = raw_data.iloc[2]
    df1_2 = df1_1.drop(['Examples','Total','PE Total','Substitute Total'], axis=1)
    
    df18 = df1_2.iloc[:,0:25]
 #   df18['Year']  = '2018'
    df18.columns =['Process','Sub Process','Market','SubMarket','Application','Unimodal HDPE',
                   'Bimodal HDPE','C4','C6','C8','_C6','_C8','VLDPE','LDPE','<7%','7%<17%','17%<25%','>25%',
                   'Bio-based PE','PBAT','PBS','PHA/PHB','PLA','Others','PE Recyclate'] 
 
    df19 = pd.concat([df1_2.iloc[:, 0:5], df1_2.iloc[:, 25:45]], axis=1)
 #   df19['Year']  = '2019'
    df19.columns =['Process','Sub Process','Market','SubMarket','Application','Unimodal HDPE',
                   'Bimodal HDPE','C4','C6','C8','_C6','_C8','VLDPE','LDPE','<7%','7%<17%','17%<25%','>25%',
                   'Bio-based PE','PBAT','PBS','PHA/PHB','PLA','Others','PE Recyclate'] 

    df20 = pd.concat([df1_2.iloc[:, 0:5], df1_2.iloc[:, 45:65]], axis=1)
 #   df20['Year']  = '2020'
    df20.columns =['Process','Sub Process','Market','SubMarket','Application','Unimodal HDPE',
                   'Bimodal HDPE','C4','C6','C8','_C6','_C8','VLDPE','LDPE','<7%','7%<17%','17%<25%','>25%',
                   'Bio-based PE','PBAT','PBS','PHA/PHB','PLA','Others','PE Recyclate']    

    df25 = pd.concat([df1_2.iloc[:, 0:5], df1_2.iloc[:, 65:85]], axis=1)
 #   df25['Year']  = '2025'
    df25.columns =['Process','Sub Process','Market','SubMarket','Application','Unimodal HDPE',
                   'Bimodal HDPE','C4','C6','C8','_C6','_C8','VLDPE','LDPE','<7%','7%<17%','17%<25%','>25%',
                   'Bio-based PE','PBAT','PBS','PHA/PHB','PLA','Others','PE Recyclate'] 
    
#    df18_d = df18.drop(['Process','Sub Process','Market','SubMarket','Application','Year'], axis=1)
#    df19_d = df19.drop(['Process','Sub Process','Market','SubMarket','Application','Year'], axis=1)
#    df20_d = df20.drop(['Process','Sub Process','Market','SubMarket','Application','Year'], axis=1)
#    df25_d = df25.drop(['Process','Sub Process','Market','SubMarket','Application','Year'], axis=1)

    unpivoted18 = df18.melt(id_vars=['Process','Sub Process','Market','SubMarket','Application'], 
                            var_name='Subresin_level2', value_name='2018')
    unpivoted19 = df19.melt(id_vars=['Process','Sub Process','Market','SubMarket','Application'], 
                            var_name='Subresin_level2', value_name='2019')
    unpivoted20 = df20.melt(id_vars=['Process','Sub Process','Market','SubMarket','Application'], 
                            var_name='Subresin_level2', value_name='2020')
    unpivoted25 = df25.melt(id_vars=['Process','Sub Process','Market','SubMarket','Application'], 
                            var_name='Subresin_level2', value_name='2025')

#    df_unpivoted = unpivoted18.append([unpivoted19, unpivoted20, unpivoted25])
    unpivoted18["2019"]=unpivoted19[["2019"]]
    unpivoted18["2020"]=unpivoted20[["2020"]]
    unpivoted18["2025"]=unpivoted25[["2025"]]
#    df_unpivoted = pd.concat([unpivoted18,unpivoted19[:,6],unpivoted20[:,6],unpivoted25[:,6],axis=1,join="inner")
    unpivoted18.reset_index(drop=True, inplace=True)
 
    unpivoted18_nan=unpivoted18.replace(0,np.nan,inplace=False)
    unpivoted18["2019r"]=pd.DataFrame(unpivoted18_nan["2019"]/unpivoted18_nan["2018"]-1)
    unpivoted18["2020r"]=pd.DataFrame(unpivoted18_nan["2020"]/unpivoted18_nan["2019"]-1)
    unpivoted18["2025r"]=pd.DataFrame((unpivoted18_nan["2025"]/unpivoted18_nan["2020"])**0.2-1)    

#    rate18 = pd.DataFrame(index=range(0,len(unpivoted18)))
#    rate18['Growth_Rate'] = 'NaN'

#    df18_d.replace(0, np.nan, inplace=True)
#    df19_d.fillna(0)
#    df1819 = df19_d - df18_d
#    rate19 = (df1819/df18_d).melt(value_name='Growth_Rate')[['Growth_Rate']]

#    df19_d.replace(0, np.nan, inplace=True)
#    df20_d.fillna(0)
#    rate20 = ((df20_d - df19_d)/df18_d).melt(value_name='Growth_Rate')[['Growth_Rate']]

#    df20_d.replace(0, np.nan, inplace=True)
#    df25_d.fillna(0)
#    rate25 = ((df25_d/df20_d) ** (1/5) -1).melt(value_name='Growth_Rate')[['Growth_Rate']]
    
#    rate = rate18.append([rate19, rate20, rate25])
#    rate.reset_index(drop=True, inplace=True)
    
#    df_unpivoted['Growth_Rate'] = rate[['Growth_Rate']]

    unpivoted18['Subresin_level1']  = 'H'
    unpivoted18['Resin']  = 'dd'
    unpivoted18['Region'] = region.split(".")[0]
    unpivoted18['Country'] = country
    unpivoted18 = unpivoted18[['Region','Country','Process','Sub Process','Market','SubMarket','Application',
                               'Resin','Subresin_level1','Subresin_level2',
                               '2018','2019','2020','2025','2019r','2020r','2025r']]

    # Subresin_level1
    unpivoted18.loc[(unpivoted18['Subresin_level2'] == 'Unimodal HDPE') | 
                    (unpivoted18['Subresin_level2'] == 'Bimodal HDPE'),'Subresin_level1'] = 'HDPE'
    unpivoted18.loc[(unpivoted18['Subresin_level2'] == 'C4') | 
                    (unpivoted18['Subresin_level2'] == 'C6') | 
                    (unpivoted18['Subresin_level2'] == 'C8'),'Subresin_level1'] = 'LLDPE'
    unpivoted18.loc[(unpivoted18['Subresin_level2'] == '_C6') | 
                    (unpivoted18['Subresin_level2'] == '_C8') ,'Subresin_level1'] = 'Metallocene'
    unpivoted18.loc[(unpivoted18['Subresin_level2'] == 'VLDPE'),'Subresin_level1'] = 'VLDPE'
    unpivoted18.loc[(unpivoted18['Subresin_level2'] == 'LDPE'),'Subresin_level1'] = 'LDPE'
    unpivoted18.loc[(unpivoted18['Subresin_level2'] == '<7%') |
                    (unpivoted18['Subresin_level2'] == '7%<17%') |
                    (unpivoted18['Subresin_level2'] == '17%<25%') |
                    (unpivoted18['Subresin_level2'] == '>25%'),'Subresin_level1'] = 'EVA'
    unpivoted18.loc[(unpivoted18['Subresin_level2'] == 'Substitute Total') |
                    (unpivoted18['Subresin_level2'] == 'Bio-based PE') |
                    (unpivoted18['Subresin_level2'] == 'PBAT') |
                    (unpivoted18['Subresin_level2'] == 'PBS') |
                    (unpivoted18['Subresin_level2'] == 'PHA/PHB') |
                    (unpivoted18['Subresin_level2'] == 'PLA') |
                    (unpivoted18['Subresin_level2'] == 'Others'),'Subresin_level1'] = 'Bioplastics'
    unpivoted18.loc[(unpivoted18['Subresin_level2'] ==
                      'PE Recyclate'),'Subresin_level1'] = 'Recyclate'

    # Resin
    unpivoted18.loc[(unpivoted18['Subresin_level1'] == 'HDPE'),'Resin'] = 'HDPE'
    unpivoted18.loc[(unpivoted18['Subresin_level1'] == 'LLDPE') |
                    (unpivoted18['Subresin_level1'] == 'Metallocene') |
                    (unpivoted18['Subresin_level1'] == 'VLDPE') ,'Resin'] = 'LLDPE'
    unpivoted18.loc[(unpivoted18['Subresin_level1'] == 'LDPE') | 
                    (unpivoted18['Subresin_level1'] == 'EVA') ,'Resin'] = 'LDPE'
    unpivoted18.loc[(unpivoted18['Subresin_level1'] == 'Bioplastics') |
                    (unpivoted18['Subresin_level1'] == 'Recyclate') ,'Resin'] = 'Substitute'
 
    unpivoted18.index.name = 'ID'
    unpivoted18.to_excel('./output/' + region.split(".")[0] + '_' + country + '.xlsx')
    if i == 0:
        unpivoted18_old = unpivoted18
    else:
        unpivoted18_old = unpivoted18_old.append(unpivoted18)
    unpivoted18_old.reset_index(drop=True, inplace=True)
    unpivoted18_old.index.name = 'ID'
 #   try:
 #       os.remove(country + '_.xlsx')
 #   except OSError:
 #       pass

unpivoted18_old.to_excel("./output/GLOBAL.xlsx")
    
```
